```python
%matplotlib inline 
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage.io import imread
```

```python
"""
This function stores the centers and 
their respective labels accordingly.
Parameters:
    data = the pixel values of the image
    k    = number of clusters 
Returns:
    None
"""
def kmeans_f(data, k):
    kmeans = KMeans(n_clusters = k, n_init = 4).fit(data)
    center = kmeans.cluster_centers_.squeeze()
    label = kmeans.labels_
    return (center, label)
```

```python
"""
This function displays the original image 
in comparison to the vector quantizated 
image.
Parameters:
    data = the pixel values of the image
    k    = number of clusters 
Returns:
    None
"""
def imagesX(img, k):
    x = img.reshape((-1, 1))
    c, l = kmeans_f(x, k)
    df = np.choose(l, c).astype('uint8')
    df.shape = img.shape
    f = plt.figure(1)
    f.set_figheight(18)
    f.set_figwidth(18)
    orig = f.add_subplot(121)
    orig.imshow(img, cmap = plt.cm.gray, vmin = (img.min()), vmax = (img.max()))
    orig.set_title('Original image')
    new = f.add_subplot(122)
    new.imshow(df, cmap = plt.cm.gray, vmin = (img.min()), vmax = (img.max()))
    new.set_title('K-Means clustered image')
    plt.show()
```

```python
"""
This function runs the K-means clustering 
on various images.
Parameters:
    None 
Returns:
    None
"""
def main():
    img = imread("d.jpg")
    imagesX(img, 2)
    img1 = imread("x.jpg")
    imagesX(img1, 2)
    img2 = imread("h1.jpg")
    imagesX(img2, 2)
    img3 = imread("h.jpg")
    imagesX(img3, 2)
    img4 = imread("y.jpg")
    imagesX(img4, 2)
    img5 = imread("y1.jpg")
    imagesX(img5, 2)
    img6 = imread("y2.jpg")
    imagesX(img6, 2)
main()
```

```python
"""
This function calculates the mean 
sum of squared error for each image
and then displays it as a graph.
Parameters:
    data = the pixel values of the image
Returns:
    None
"""
def sse(data):
    x = data.reshape((-1, 1))
    Error =[]
    for i in range(1, 11):
        kmeans = KMeans(n_clusters = i).fit(x)
        kmeans.fit(x)
        Error.append(kmeans.inertia_)
    plt.plot(range(1, 11), Error)
    plt.title('Error')
    plt.xlabel('k value')
    plt.ylabel('Mean Sum of Squares')
    plt.show()
```
